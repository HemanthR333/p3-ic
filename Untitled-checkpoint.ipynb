```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential 
```

```python
data_train_path =  'Fruits_Vegetables/Fruits_Vegetables/train'
data_val_path = 'Fruits_Vegetables/Fruits_Vegetables/validation'
data_test_path = 'Fruits_Vegetables/Fruits_Vegetables/test'
```

```python
img_width , img_height = 180,180
```

```python
data_train = tf.keras.utils.image_dataset_from_directory(
    data_train_path,
    shuffle = True,
    image_size = (img_width , img_height),
    batch_size = 32,
    validation_split = False
)
```

```python
data_val = tf.keras.utils.image_dataset_from_directory(
    data_val_path,
    shuffle = False,
    batch_size = 32,
    image_size = (img_width , img_height),
    validation_split = False
)
```

```python
data_test = tf.keras.utils.image_dataset_from_directory(
    data_test_path,
    shuffle = False,
    batch_size = 32,
    image_size = (img_width , img_height),
    validation_split = False
)
```

```python
data_cat = data_train.class_names 
```

```python
plt.figure(figsize = (10,10))
for  image , labels in data_train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(data_cat[labels[i]])
        plt.axis('off')
```

```python
model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3 , padding = 'same', activation = 'relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3 , padding = 'same' , activation = 'relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64 , 3, padding = 'same' , activation = 'relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128),
    layers.Dense(len(data_cat))
])
```

```python
model.compile(optimizer = 'adam' , loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
```

```python
history = model.fit(data_train , validation_data = data_val , epochs = 10)
```

```python
epoch_size = range(10)
plt.figure(figsize= (8,8))
plt.subplot(1,2,1)
plt.plot(epoch_size , history.history['accuracy'], label = 'Training Accuracy')
plt.plot(epoch_size , history.history['val_accuracy'] , label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epoch_size , history.history['loss'] , label = 'Training Loss')
plt.plot(epoch_size , history.history['val_loss'] , label = 'Validation Loss')
plt.title('Loss')
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import load_model

model = load_model('image_detection.h5')
labels = ['apple',
 'banana',
 'beetroot',
 'bell pepper',
 'butter',
 'cabbage',
 'capsicum',
 'carrot',
 'cauliflower',
 'cheese',
 'chilli pepper',
 'cocacola',
 'corn',
 'cream',
 'cucumber',
 'eggplant',
 'garlic',
 'ginger',
 'grapes',
 'ice cream',
 'jalepeno',
 'kiwi',
 'lemon',
 'lettuce',
 'mango',
 'milk',
 'onion',
 'orange',
 'paprika',
 'pear',
 'peas',
 'pineapple',
 'pomegranate',
 'potato',
 'raddish',
 'soy beans',
 'spinach',
 'sweetcorn',
 'sweetpotato',
 'tomato',
 'turnip',
 'watermelon']
cam = cv2.VideoCapture(0)

objects = []

while True:
    ret , frame = cam.read()
    #img = cv2.rectangle(frame,(20,80),(550,350),(0,0,0),2)
    image = cv2.resize(frame, (180, 180), interpolation=cv2.INTER_AREA)
    cv2.imshow('frame',image)
    image = np.asarray(image, dtype=np.float32).reshape(1, 180, 180, 3)
    image = (image / 127.5) - 1
    if cv2.waitKey(1)== ord('c'):
        score =tf.nn.softmax(model.predict(image)) 
        objects.append(labels[np.argmax(score)])
    elif cv2.waitKey(1) == ord('q'):
        break
print(objects)
cam.release()
cv2.destroyAllWindows()
```
